```python
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
from sklearn import decomposition
```

```python
d = datasets.load_iris()
```

```python
data = pd.DataFrame(d['data'] , columns= d['feature_names'])
pca = decomposition.PCA(3)
data = pca.fit_transform(data)
```

```python
data = np.c_[data,d['target']]
df = pd.DataFrame(data, columns= ['PC1', 'PC2', 'PC3', 'TARGET'])
```

```python
df.head()
```

```python
fig = plt.figure(figsize = (15,10))
ax = Axes3D(fig)
plot = ax.scatter3D(df.PC1, df.PC2, df.PC3 , c = df.TARGET, s = 300 )
ax.set_xlabel('PC1', fontsize = 20)
ax.set_ylabel('PC2', fontsize = 20)
ax.set_zlabel('PC3', fontsize = 20)
ax.set_title('Scatter 3d plot of iris dataset', fontsize = 20)
#plt.colorbar(plot, ax = ax)
plt.legend(*plot.legend_elements(), fontsize = 20)
plt.show()
```
